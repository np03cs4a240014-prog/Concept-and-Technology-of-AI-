# b. Composite Development Score:

- Create a new metric called Composite Score using the formula:

Composite Score = 0.30 × Life Expectancy Index + 0.30 × GNI per Capita Index}
Here: Life Expectancy Index → "life expectancy" and GNI per Capita Index → "gross inc percap"

```python
hdi_south_asian_countries= pd.read_csv('/content/drive/MyDrive/2501467_SubasanaPrabhaBajracharya/HDI_SouthAsia.csv')
latest_year_df = hdi_south_asian_countries.sort_values('year').groupby('country').tail(1)
latest_year_df['life_exp_index'] = (latest_year_df['life_expectancy'] - latest_year_df['life_expectancy'].min()) / (latest_year_df['life_expectancy'].max() - latest_year_df['life_expectancy'].min())

latest_year_df['gni_index'] = (latest_year_df['gross_inc_percap'] - latest_year_df['gross_inc_percap'].min()) / (latest_year_df['gross_inc_percap'].max() - latest_year_df['gross_inc_percap'].min())

latest_year_df['composite_score'] = 0.3 * latest_year_df['life_exp_index'] + 0.3 * latest_year_df['gni_index']


```

- Rank South Asian countries based on Composite Score.

```python
ranked_df = latest_year_df.sort_values('composite_score', ascending=False)
print("Ranking by Composite Score:")
print(ranked_df[['country', 'composite_score', 'hdi']])
```

- Plot the top 5 countries in a horizontal bar chart.

```python
top_5_countries = ranked_df.head(5)
plt.figure(figsize=(10,6))
plt.barh(top_5_countries['country'], top_5_countries['composite_score'], color='skyblue')
plt.xlabel('Composite Score')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.gca().invert_yaxis() #represents the highest on top
plt.show()
```

- Compare the ranking of countries by Composite Score with their HDI ranking and discuss any
differences.

```python

hdi_ranked = latest_year_df.sort_values('hdi', ascending=False).reset_index(drop=True)
hdi_ranked['hdi_rank'] = hdi_ranked.index + 1
composite_ranked = ranked_df.reset_index(drop=True)
composite_ranked['composite_rank'] = composite_ranked.index + 1

comparison_df = pd.merge(
    hdi_ranked[['country', 'hdi', 'hdi_rank']],
    composite_ranked[['country', 'composite_score', 'composite_rank']],
    on='country'
)

print(comparison_df)
```

The comparison table compares countries based on two different ranking methods: HDI rank and composite rank. The HDI rank is derived solely from the Human Development Index (HDI) value, whereas the composite rank is based on a combined score that includes multiple indicators.

# 3. Outlier Detection:

- Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.

```python
Q1_score=latest_year_df['hdi'].quantile(0.25)
Q3_score=latest_year_df['hdi'].quantile(0.75)
IQR_score=Q3_score-Q1_score

IQR_score-Q3_score-Q1_score

Q1_gni=latest_year_df['gross_inc_percap'].quantile(0.25)
Q3_gni=latest_year_df['gross_inc_percap'].quantile(0.75)

IQR_gni=Q3_gni-Q1_gni

hdi_outliners = latest_year_df[(latest_year_df['hdi'] < Q1_score - 1.5 * IQR_score) |
                              (latest_year_df['hdi'] > Q3_score + 1.5 * IQR_score)]
gni_outliners = latest_year_df[(latest_year_df['gross_inc_percap'] < Q1_gni - 1.5 * IQR_gni) |
                              (latest_year_df['gross_inc_percap'] > Q3_gni + 1.5 * IQR_gni)]

outliners =pd.concat([hdi_outliners, gni_outliners]).drop_duplicates()
display(outliners)
```

- Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

```python

plt.scatter(latest_year_df['gross_inc_percap'], latest_year_df['hdi'], color='black', label='Normal')
plt.scatter(outliners['gross_inc_percap'], outliners['hdi'], color='red', label='Outliers')

plt.title('GNI per Capita vs HDI')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.legend()
plt.show()
```

# 7. Outlier Detection:

- Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

```python
#Consists for south asian countries
Q1_hdi_south_asia = south_asia['hdi'].quantile(0.25)
Q3_hdi_south_asia = south_asia['hdi'].quantile(0.75)
IQR_hdi_south_asia = Q3_hdi_south_asia - Q1_hdi_south_asia

Q1_gni_south_asia = south_asia['gross_inc_percap'].quantile(0.25)
Q3_gni_south_asia = south_asia['gross_inc_percap'].quantile(0.75)
IQR_gni_south_asia = Q3_gni_south_asia - Q1_gni_south_asia

hdi_outliners_south_asia = south_asia[(south_asia['hdi'] < Q1_hdi_south_asia - 1.5 * IQR_hdi_south_asia) | (south_asia['hdi'] > Q3_hdi_south_asia + 1.5 * IQR_hdi_south_asia)]
gni_outliners_south_asia = south_asia[(south_asia['gross_inc_percap'] < Q1_gni_south_asia - 1.5 * IQR_gni_south_asia) | (south_asia['gross_inc_percap'] > Q3_gni_south_asia + 1.5 * IQR_gni_south_asia)]

outliners_south_asia = pd.concat([hdi_outliners_south_asia, gni_outliners_south_asia])
outliners_south_asia = outliners_south_asia.drop_duplicates()

print("Outliners in South Asia:")
display(outliners_south_asia)

```

```python
#Consists for south asian countries
Q1_hdi_middle_east = middle_east['hdi'].quantile(0.25)
Q3_hdi_middle_east = middle_east['hdi'].quantile(0.75)
IQR_hdi_middle_east = Q3_hdi_middle_east - Q1_hdi_middle_east

Q1_gni_middle_east = middle_east['gross_inc_percap'].quantile(0.25)
Q3_gni_middle_east = middle_east['gross_inc_percap'].quantile(0.75)
IQR_gni_middle_east = Q3_gni_middle_east - Q1_gni_middle_east

hdi_outliners_middle_east = middle_east[(middle_east['hdi'] < Q1_hdi_middle_east - 1.5 * IQR_hdi_middle_east) | (middle_east['hdi'] > Q3_hdi_middle_east + 1.5 * IQR_hdi_middle_east)]
gni_outliners_middle_east = middle_east[(middle_east['gross_inc_percap'] < Q1_gni_middle_east - 1.5 * IQR_gni_middle_east) | (middle_east['gross_inc_percap'] > Q3_gni_middle_east + 1.5 * IQR_gni_middle_east)]

outliners_middle_east = pd.concat([hdi_outliners_middle_east, gni_outliners_middle_east])
outliners_middle_east = outliners_middle_east.drop_duplicates()

print("Outliners in Middle East:")
display(outliners_middle_east)

```
